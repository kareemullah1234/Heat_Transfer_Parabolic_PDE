```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] ='serif'
pyplot.rcParams['font.size'] = 16
```

```python
L =1.0
nx = 11
dx = L/ (nx-1) # why (n-1) because distance b/w two points will be n-1 for n points
alpha = 1.22e-3 

x =numpy.linspace(0.0,L,num=nx)

T0  =numpy.zeros(nx)
T0[0] =100.0
r=[]
```

```python
def ftcs(T0,nt,dt,dx,alpha): #nt,dt is not defined
    T =T0.copy()
    sigma = alpha * dt / dx**2 #dt is not defined
    
    #print(r)
    for n in range(nt):
        
        
        T[1:-1] = (T[1:-1] +  sigma * (T[2:] - 2* T[1:-1]+ T[:-2]))
        
        
        r.append(n)
    #print(r)
    return T
   
```

```python
nt = 11
sigma = 0.5
dt =sigma * dx**2 / alpha

T = ftcs(T0,nt,dt,dx,alpha)
r
```

```python
pyplot.figure(figsize=(6.0,4.0))
pyplot.xlabel('Distance[m] and time')
pyplot.ylabel('Temperature[c]')
pyplot.grid()
pyplot.plot(x,T,color='C0',linestyle='-',linewidth=2)
pyplot.plot(r,T,color='C0',linestyle='-',linewidth=2)
pyplot.xlim(0.0,L)
pyplot.ylim(0.0,100.0)
```
